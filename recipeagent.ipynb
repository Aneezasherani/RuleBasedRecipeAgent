```python
import pandas as pd
```

```python
df = pd.read_csv('recipes.csv')
```

```python
print("Sample recipes dataset:")
print(df.head())
```

```python
def recommend_dish_subset(available_ingredients):
    recommendations = []

    for index, row in df.iterrows():
        recipe_ingredients = [i.strip().lower() for i in row['Ingredients'].split(",")]

        # Check if all user ingredients are in recipe ingredients
        if all(item in recipe_ingredients for item in available_ingredients):
            recommendations.append({
                "Recipe Name": row['Recipe Name'],
                "Steps": row['Steps'],
                "Recipe Ingredients": recipe_ingredients
            })

    return recommendations
```

```python
user_input = input("Enter available ingredients (comma separated): ")
available_ingredients = [item.strip().lower() for item in user_input.split(",")]

# Run agent
recommended = recommend_dish_subset(available_ingredients)

# Show results
if recommended:
    print("\nRecipes you can make with your ingredients:\n")
    for r in recommended:
        print("Dish:", r['Recipe Name'])
        print("Recipe Ingredients:", ", ".join(r['Recipe Ingredients']))
        print("Steps:\n", r['Steps'])
        print("-"*50)
else:
    print("\nNo recipe found with all given ingredients!")
```
